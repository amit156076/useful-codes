```python
#get image
# 1143 x 1390
mp = Image.open("/home/amit/Pictures/mpmap1.jpg")
```

```python
rgb = mp.convert('RGB')
#r value for mp area 227 r value for other map area 186
```

```python
count = 0
count_in = 0
count_mp = 0
```

```python
#area of madhya pradesh
while count < 1000000:
    x = np.random.randint(0, 1143)
    y = np.random.randint(0, 1390)
    r,g,b = rgb.getpixel((x, y))
    if g < 100:
        count_mp = count_mp + 1
        count = count + 1
    elif g > 100 & g < 255:
        count_in = count_in + 1
        count = count + 1
print("area of mp:", (count_mp/count_in)* 3287000)
print("india count:", count_in)
print("mp count:", count_mp)
```

```python
#image size 244 x 280
upimg = Image.open("/home/amit/Pictures/upmap1.jpg")
```

```python
rgb1 = upimg.convert('RGB')
```

```python
count = 0
count_in = 0
count_up = 0
```

```python
#area of uttar pradesh
while count < 10000:
    x = np.random.randint(0, 244)
    y = np.random.randint(0, 280)
    r,g,b = rgb1.getpixel((x, y))
    if b == 0:
        count_up = count_up + 1
        count = count + 1
    elif b > 100 & b < 150 :
        count_in = count_in + 1
        count = count + 1
print("area of up:", (count_up/count_in)* 3287000)
print("india count:", count_in)
print("up count:", count_up)
print("fraction:", count_up/count_in)
```
