```python
import numpy as np
```

```python
from PIL import Image
```

```python
#get image
mp = Image.open("/home/amit/Pictures/maps/mpmap2.jpg")
```

```python
rgb = mp.convert('RGB')
```

```python
count = 0
count_in = 0
count_mp = 0
```

```python
#area of madhya pradesh
while count < 10000:
    x = np.random.randint(0, rgb.size[0])
    y = np.random.randint(0, rgb.size[1])
    r,g,b = rgb.getpixel((x, y))
 #the logic to extract the are based on r, g, b values. It will be different for every image and scenario.
 #to know the r g b values of any point in image you can use website https://www.ginifab.com/feeds/pms/color_picker_from_image.php
    if r < 255:
        if r == 177:
            count_mp = count_mp + 1
        else:
            count_in = count_in + 1
        count = count + 1
      
print("area of mp:", (count_mp/count_in)* 3287000)
print("india count:", count_in)
print("mp count:", count_mp)
```

```python
#image size 244 x 280
upimg = Image.open("/home/amit/Pictures/maps/upmap1.jpg")
```

```python
rgb1 = upimg.convert('RGB')
```

```python
count = 0
count_in = 0
count_up = 0
```

```python
#area of uttar pradesh
while count < 10000:
    x = np.random.randint(0, rgb1.size[0])
    y = np.random.randint(0, rgb1.size[1])
    r,g,b = rgb1.getpixel((x, y))
    #the logic to extract the are based on r, g, b values. It will be different for every image and scenario.
    #to know the r g b values of any point in image you can use website https://www.ginifab.com/feeds/pms/color_picker_from_image.php
    if r+g+b < 302:
        count_up = count_up + 1
    elif r+g+b > 562:
        count_in = count_in + 1
    
    count = count + 1
print("area of up:", (count_up/count_in)* 3287000)
print("india count:", count_in)
print("up count:", count_up)
print("fraction:", count_up/count_in)
```
