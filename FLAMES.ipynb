```python
#convert name string to list
def str_to_list(name1, name2):
    list1 = []
    list2 = []
    for i in name1:
        list1.append(i)
    for i in name2:
        list2.append(i)
    return list1, list2
 
```

```python
#delete common letters
def del_common_letters(list1, list2):
    list1_del = []
    for i in range(0, len(list1)):
        for j in range(0, len(list2)):
            if list1[i] == list2[j]:
                list2.pop(j)
                list1_del.append(list1[i])
                break
    for i in list1_del:
        list1.remove(i)
  
    #return list1, list2
             
```

```python
def calculate_flames_value(list1, list2, flames):
    count = len(list1)+len(list2)
    while len(flames) > 1:
        i = count
        if i > len(flames):
            while i  > len(flames):
                i = i  - len(flames)

        flames.pop(i-1)
        for j in range(0, i-1):
            flames.append(flames.pop(0))
        
```

```python
def flames_game(name1, name2):
    fl = {"F":"Friends", "L":"Lovers", "A":"Affectionate", "M":"Marriage", "E":"Enemy", "S":"Siblings"}
    flames = ['F', 'L', 'A', 'M', 'E', 'S']
    
    #convert name string to list
    
    list1, list2 = str_to_list(name1, name2)
    
    #delete common letters from lists
    
    del_common_letters(list1, list2)
    
    #calculate flames value
    
    calculate_flames_value(list1, list2, flames)
    
    return fl[flames[0]]
    
      

    
```
